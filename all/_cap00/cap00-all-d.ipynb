## 📘 Parte II – Visão Computacional (Capítulos 6–10)

### **Capítulo 6 – Reconhecimento de Padrões**
- Aplicações: OCR, reconhecimento de placas de veículos, biometria simples.  
- Exercício prático: usar **OpenCV + Tesseract** para extrair texto de imagens sintéticas (placas, documentos).  
- Dataset: imagens geradas de placas ou textos impressos.  

---

### **Capítulo 7 – Detecção de Objetos**
- Aplicações: vigilância, carros autônomos, análise esportiva.  
- Exercício prático: usar **YOLOv5 pré-treinado** em imagens do COCO dataset (livres).  
- Mostrar como detectar múltiplos objetos em cena e interpretar bounding boxes.  

---

### **Capítulo 8 – Segmentação de Imagens**
- Aplicações: medicina (mamografia, microscopia), agricultura (contagem de plantas), indústria (inspeção).  
- Exercício prático: usar **U-Net** para segmentar células em imagens de microscopia (dataset público).  
- Comparar segmentação clássica (Watershed, limiarização) vs. segmentação profunda.  

---

### **Capítulo 9 – Reconstrução 3D e Movimento**
- Aplicações: robótica, realidade aumentada, reconstrução de ambientes.  
- Exercício prático: reconstrução de cena com pares estéreo (Middlebury dataset) e Optical Flow em vídeos curtos.  
- Mostrar como extrair profundidade e movimento.  

---

### **Capítulo 10 – Aplicações Avançadas e Projeto Final**
- Aplicações integradas:  
  - Reconhecimento facial (FaceNet ou dlib).  
  - Rastreamento de pessoas em vídeo (DeepSORT).  
  - Classificação de imagens médicas (transfer learning com ResNet).  
- Exercício prático: mini-projeto final com escolha livre (OCR, detecção de objetos, segmentação, reconstrução 3D).  

---

## 🎯 Estratégia
- **Pouca teoria**: apenas o essencial para contextualizar.  
- **Muitos exemplos práticos**: cada capítulo com 2–3 notebooks aplicados.  
- **Datasets públicos e sintéticos**: MNIST, CIFAR-10, COCO, Middlebury, Open Images, ou imagens geradas (xadrez, sudoku, figuras geométricas).  
- **Aplicações reais**: sempre mostrar como aquilo é usado em medicina, indústria, robótica, etc.  
- **Exercícios incrementais**: cada notebook termina com desafios práticos para os alunos adaptarem.
