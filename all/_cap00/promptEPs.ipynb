    "### EP03_01 ☀️ Ajuste de Brilho e Contraste\n",
    "\n",
    "Nesta atividade, o objetivo é implementar um operador de ponto para **transformação linear de intensidade**, aplicando o ajuste dinâmico de brilho e contraste em uma imagem digital.\n",
    "\n",
    "#### 📋 Diretrizes de Implementação\n",
    "\n",
    "O algoritmo deve seguir o fluxo de execução abaixo:\n",
    "\n",
    "1. **Dimensões:** Ler os inteiros $L$ (linhas) e $C$ (colunas) da matriz.\n",
    "2. **Parâmetros:** Ler o valor real $\\alpha$ (fator de contraste) e o inteiro $\\beta$ (fator de brilho).\n",
    "3. **Dados:** Ler os valores inteiros da matriz original.\n",
    "4. **Mapeamento:** Para cada pixel $p$, calcular o novo valor $p'$ pela equação:\n",
    "\n",
    "$$p' = \\text{clip}(\\text{round}(\\alpha \\cdot p + \\beta))$$\n",
    "\n",
    "5. **Saída:** Exibir a matriz resultante com as dimensões $L \\times C$.\n",
    "\n",
    "---\n",
    "\n",
    "#### 📌 Restrições Computacionais\n",
    "\n",
    "* **Arredondamento (*Round*):** Aplica-se o arredondamento matemático para o inteiro mais próximo antes da conversão de tipo.\n",
    "* **Saturação (*Clipping*):** Os valores devem ser confinados ao intervalo $[0, 255]$ para preservar o padrão de 8 bits:\n",
    "\n",
    "$$\\text{clip}(x) = \\max(0, \\min(255, x))$$\n",
    "\n",
    "* **Simulação:** O efeito dos parâmetros $\\alpha$ e $\\beta$ na correção histogramática pode ser observado na @fig-03-bril-cont.\n",
    "\n",
    "---\n",
    "\n",
    "#### 🧠 Fundamentação Teórica\n",
    "\n",
    "As alterações modificam o histograma da imagem para ajustar o perfil de iluminação e a distinção tonal.\n",
    "\n",
    "| Parâmetro | Função | Impacto Visual |\n",
    "| --- | --- | --- |\n",
    "| **$\\alpha$ (*Alpha*)** | Escalar | Modula o **Contraste**. Se $\\alpha > 1$, expande o histograma; se $0 \\le \\alpha < 1$, comprime. |\n",
