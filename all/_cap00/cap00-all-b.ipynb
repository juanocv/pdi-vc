# Estrutura que eu realmente adotaria

Se o objetivo é um livro **extremamente didático e motivador**, eu faria:

| Capítulo | Pergunta                        |
| -------- | ------------------------------- |
| 6        | Onde está o objeto?             |
| 7        | O que é o objeto?               |
| 8        | O objeto está se movendo?       |
| 9        | Qual a distância até o objeto?  |
| 10       | Como a IA moderna resolve isso? |

Observe que o aluno passa naturalmente por:

```text
Detectar
↓
Classificar
↓
Rastrear
↓
Medir
↓
Usar IA
```

que é exatamente o fluxo mental de praticamente qualquer sistema moderno de Visão Computacional.

Além disso, essa sequência permite reutilizar os mesmos conjuntos de imagens ao longo dos capítulos, reduzindo bastante o trabalho de produção do livro e dos notebooks. Por exemplo, um conjunto de imagens de veículos pode ser usado para:

* segmentação (Cap. 5)
* detecção (Cap. 6)
* classificação (Cap. 7)
* rastreamento (Cap. 8)
* estimativa de velocidade/distância (Cap. 9)
* YOLO (Cap. 10)

Isso gera uma sensação de continuidade que raramente aparece em livros de PDI/VC.
