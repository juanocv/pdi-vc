### Parte II: Visão Computacional (Foco Total em Aplicações)

#### **Capítulo 6: Inspeção Industrial e Análise de Documentos (Visão 2D Clássica)**

* **Foco da Aplicação:** Construir um sistema automatizado para controle de qualidade em esteiras e preparação de documentos para leitura.
* **Projetos Práticos no Notebook:**
1. **Contagem e Inspeção de Peças Industriais:** Usar binarização adaptativa, detecção de contornos e momentos geométricos para contar parafusos/engrenagens em uma imagem e rejeitar peças com defeito (área ou perímetro incorretos).
2. **Alinhamento Automático de Documentos (OCR Pre-processing):** Detectar as bordas de uma folha de papel digitalizada torta usando a *Transformada de Hough*, calcular o ângulo de inclinação e rotacionar a imagem automaticamente para deixar o texto perfeitamente horizontal.


* **Teoria Enxuta:** Apenas a intuição de como os contornos organizam os pixels em estruturas vetoriais.

#### **Capítulo 7: Biometria e Rastreamento de Objetos (Extração de Features e Vídeo)**

* **Foco da Aplicação:** Trabalhar com vídeo real, lidando com análise espaço-temporal e identificação de padrões baseados em textura/forma.
* **Projetos Práticos no Notebook:**
1. **Reconhecimento de Faces com Eigenfaces/Fisherfaces:** Implementar o pipeline clássico de biometria facial. Reduzir a dimensionalidade de um banco de faces usando PCA (Análise de Componentes Principais) e classificar usando uma distância simples (K-NN), conectando direto com os EPs de métricas que você já tem.


2. **Rastreamento de Tráfego por Fluxo Óptico:** Analisar um vídeo de câmera de segurança viária. Usar o algoritmo de Lucas-Kanade ou segmentação de fundo (*Background Subtraction*) para detectar carros em movimento e desenhar seus vetores de velocidade na tela.


* **Teoria Enxuta:** Como vetores de características (*feature vectors*) resumem uma imagem ou representam o movimento entre dois frames consecutivos.

#### **Capítulo 8: Visão Robótica e Reconstrução (Aplicações 3D Descomplicadas)**

* **Foco da Aplicação:** Fazer o computador entender distância e perspectiva usando ferramentas prontas de calibração, sem deduzir equações de geometria projetiva.
* **Projetos Práticos no Notebook:**
1. **Escaneamento e Medição de Objetos Reais:** Utilizar fotos de um tabuleiro de xadrez para rodar a calibração da câmera (`cv2.calibrateCamera`). Com os parâmetros calculados, medir o tamanho real (em centímetros) de um objeto posicionado na mesa.
2. **Geração de Mapas de Profundidade para Robôs:** Pegar duas imagens de um sistema estéreo (câmera esquerda e direita) e usar o bloco `cv2.StereoBM` para gerar um mapa de disparidade em tons de cinza, mostrando o que está perto (mais claro) e o que está longe (mais escuro).


* **Teoria Enxuta:** O conceito visual de paralaxe (o mesmo do olho humano) e como uma matriz de calibração converte pixels em milímetros.

#### **Capítulo 9: Detecção de Objetos com Deep Learning (Pipelines Modernos com YOLO)**

* **Foco da Aplicação:** Utilizar o estado da arte em redes neurais profundas de forma direta para resolver problemas complexos de contagem e localização.
* **Projetos Práticos no Notebook:**
1. **Contador de Ocupação de Vagas em Estacionamento:** Usar o modelo YOLO pré-treinado em pouquíssimas linhas de código (via biblioteca `ultralytics`) para detectar carros em um vídeo de estacionamento e programar uma lógica que conta quantas vagas estão preenchidas.
2. **Leitura Automatizada de Placas de Veículos (ALPR):** Criar um pipeline de duas etapas: uma rede YOLO pequena localiza a placa do carro no frame; a região da placa é recortada e passada para uma biblioteca de OCR (como `EasyOCR` ou `Tesseract`) para extrair o texto alfanumérico.


* **Teoria Enxuta:** A diferença prática entre Classificação (o que é?), Detecção (onde está?) e Segmentação Semântica (quais pixels pertencem ao objeto?).

#### **Capítulo 10: Visão Computacional Aplicada à Saúde (Grandes Desafios)**

* **Foco da Aplicação:** Processamento de imagens biomédicas (mamografia, microscopia), conectando com a ementa da pós e com sua especialidade de pesquisa.
* **Projetos Práticos no Notebook:**
1. **Triagem Automatizada de Mamografias:** Pegar imagens médicas e usar *Transfer Learning* (uma rede pré-treinada como ResNet ou MobileNet no PyTorch/Keras) congelando quase todas as camadas, apenas tunando a última camada para classificar o tecido como "Normal" ou "Alterado".
2. **Segmentação de Células em Microscopia:** Implementar uma rede estilo U-Net simplificada (ou usar modelos prontos como Cellpose/Segment Anything via API) para segmentar e separar células sobrepostas em imagens de biologia, calculando automaticamente a densidade celular da amostra.


* **Teoria Enxuta:** Como o *Transfer Learning* permite usar o conhecimento de redes gigantes treinadas na ImageNet para resolver problemas médicos com pouquíssimas imagens de treino.

### O que muda nos Exercícios de Programação (EPs)?

Com essa abordagem focada em aplicações, os EPs da Parte II deixam de ser "implemente a fórmula X" e passam a ser **"complete o pipeline do projeto"**:

* O aluno recebe o esqueleto do código carregando o vídeo ou as imagens do dataset público (KITTI, MIAS, etc.).
* O EP desafia o aluno a: tunar os hiperparâmetros (ajustar limiares de Canny, mudar o tamanho do kernel, alterar a confiança do YOLO) e escrever as funções lógicas de tomada de decisão (ex: "se a área do contorno for menor que 100, descarte; se houver um carro cruzando a linha Y, incremente o contador").
* A validação automática foca na saída do pipeline (ex: verificar se o script do aluno acertou a quantidade de objetos ou se atingiu o mAP esperado no mini-dataset).



Essa estrutura vai manter o livro **extremamente motivante**, prático e com cara de "computação moderna", acelerando o aprendizado tanto dos alunos de graduação (que adoram ver o código funcionando em tempo real) quanto de pós-graduação (que aprendem a montar pipelines para seus experimentos de pesquisa).
