## Transformada de Fourier

### Frequência Espacial

Em imagens, o conceito de frequência está associado à velocidade com que os níveis de intensidade variam espacialmente.

* **Baixas frequências** correspondem a regiões suaves e estruturas globais;
* **Altas frequências** correspondem a mudanças abruptas, como bordas, detalhes finos e ruído.

Uma imagem pode ser interpretada como a soma de diferentes componentes de frequência distribuídas ao longo do espaço.

### Intuição Visual

Antes da formulação matemática, é importante compreender intuitivamente o significado de frequência espacial. Padrões largos e suaves produzem baixas frequências, enquanto padrões alternados e abruptos produzem altas frequências.

A figura introdutória deve mostrar:

* gradientes suaves;
* listras largas;
* listras finas;
* ruído periódico.

---

### Transformada Discreta de Fourier 2D (DFT)

A Transformada Discreta de Fourier bidimensional converte uma imagem do domínio espacial para o domínio da frequência:

genui{"math_block_widget_always_prefetch_v2":{"content":"F(u,v)=\sum_{x=0}^{M-1}\sum_{y=0}^{N-1}f(x,y)e^{-j2\pi(\frac{ux}{M}+\frac{vy}{N})}"}}

onde:

* $f(x,y)$ representa a imagem original;
* $F(u,v)$ representa a representação espectral;
* $(u,v)$ correspondem às frequências espaciais.

A DFT produz valores complexos, normalmente analisados por meio de:

* magnitude;
* fase.

A magnitude informa a intensidade das frequências presentes. A fase preserva a organização estrutural da imagem.

---

### Magnitude e Fase

Uma observação importante em Processamento Digital de Imagens é que a fase frequentemente carrega mais informação estrutural do que a magnitude.

A seção deve demonstrar:

* reconstrução usando apenas magnitude;
* reconstrução usando apenas fase;
* comparação visual.

O objetivo didático é mostrar que a fase preserva geometria e organização espacial dos objetos.

---

### Espectro Centralizado

Por convenção, o espectro costuma ser centralizado para facilitar interpretação visual:

```python
F_shift = np.fft.fftshift(F)
```

Após o deslocamento:

* baixas frequências ficam no centro;
* altas frequências ficam nas extremidades.

A visualização normalmente utiliza escala logarítmica:

```python
magnitude = np.log(1 + np.abs(F_shift))
```

## Teorema da Convolução

O Teorema da Convolução estabelece uma relação fundamental entre domínio espacial e domínio da frequência:

genui{"math_block_widget_always_prefetch_v2":{"content":"f(x,y)*h(x,y)\leftrightarrow F(u,v)H(u,v)"}}

Isso significa que:

* convolução espacial ↔ multiplicação em frequência;
* multiplicação espacial ↔ convolução em frequência.

Esse resultado explica matematicamente os filtros estudados no Capítulo 3.

---

## Wavelets e Multirresolução

* escala;

## Compressão de Imagens

### Motivação

Imagens digitais possuem elevada redundância:

* pixels vizinhos semelhantes;
* regiões homogêneas;
* padrões repetitivos.

Compressão busca reduzir armazenamento e transmissão mantendo qualidade visual aceitável.

---

### Tipos de Redundância

| Tipo        | Descrição                      |
| ----------- | ------------------------------ |
| Espacial    | pixels vizinhos semelhantes    |
| Estatística | símbolos frequentes            |
| Psicovisual | limitações da percepção humana |

---

### Compressão com Perda e Sem Perda

| Tipo      | Característica            |
| --------- | ------------------------- |
| Sem perda | reconstrução exata        |
| Com perda | pequena degradação visual |

Relacionar:

* PNG → sem perda;
* JPEG → com perda.

---

### Transformada DCT e JPEG

A compressão JPEG utiliza a Transformada Discreta do Cosseno (DCT).

## Comparação de Formatos

### JPEG

Características:

* compressão com perda;
* excelente para fotografias;
* baixa taxa de armazenamento.

Limitações:

* artefatos em compressões elevadas;
* inadequado para imagens binárias e texto.

---

### PNG

Características:

* compressão sem perda;
* preserva bordas nítidas;
* suporta transparência.

Aplicações:

* gráficos;
* diagramas;
* imagens médicas;
* interfaces.

---

### WebP

Apresentar como formato moderno híbrido:

* suporta compressão com e sem perda;
* melhor eficiência média;
* amplamente utilizado na Web.

---
