## 3.1 Processamento de Histograma

O histograma de uma imagem digital com níveis de intensidade no intervalo $[0, L-1]$ é uma função discreta $h(r_k) = n_k$, onde $n_k$ é o número de pixels com intensidade $r_k$. 

### 3.1.1 Equalização de Histograma

A equalização é uma técnica que utiliza uma transformação baseada na Função de Distribuição Acumulada (FDA) para "achatar" o histograma, distribuindo as intensidades de forma mais uniforme. Isso resulta em um aumento de contraste global, útil em imagens médicas ou capturas com iluminação deficiente.

## 3.2 Filtragem Espacial e Convolução

A filtragem espacial opera em uma vizinhança de pixels. O processo central é a **convolução**, onde uma pequena matriz (máscara ou *kernel*) desliza sobre a imagem.

### 3.2.1 Filtros de Suavização (Passa-Baixa)
Reduzem o ruído e detalhes finos. O filtro de **Média** calcula a média aritmética da vizinhança, enquanto o **Gaussiano** atribui pesos maiores aos pixels centrais, preservando melhor as bordas.

### 3.2.2 Filtros de Nitidez (Passa-Alta)
Realçam transições de intensidade (bordas). O **Laplaciano** é um operador derivativo de segunda ordem muito utilizado para destacar detalhes finos que foram atenuados.


## 3.3 Morfologia Matemática

Baseada na teoria dos conjuntos, a morfologia processa imagens de acordo com sua forma. Utiliza um **elemento estruturante** (EE) para investigar a estrutura da imagem.

<a id="tbl-03-morfologia"></a>

**Tabela 3.1:** Operações morfológicas básicas e seus efeitos.

| Operação | Descrição | Aplicação Típica |
| :--- | :--- | :--- |
| **Erosão** | "Encolhe" os objetos. | Remove ruídos pequenos (isolas). |
| **Dilatação** | "Expande" os objetos. | Preenche buracos e conecta componentes. |
| **Abertura** | Erosão seguida de dilatação. | Suaviza contornos e remove saliências. |
| **Fechamento** | Dilatação seguida de erosão. | Elimina pequenos buracos e fendas. |


```python
import cv2
import numpy as np
from morph import mm

img = mm.read("https://raw.githubusercontent.com/fzampirolli/pdi-vc/master/figures/lena.png", gray=True)

# Aplicação de filtros
blur = cv2.GaussianBlur(img, (5,5), 0)
laplacian = cv2.Laplacian(img, cv2.CV_64F).astype(np.uint8)

mm.show([img, blur, laplacian], titles=["Original", "Gaussiano (Suavização)", "Laplaciano (Bordas)"])
```
